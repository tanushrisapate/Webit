```python
#converting this to an array of just keywords
#['Action','Adventure','Fantasy']
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    list1 = []
    for i in ast.literal_eval(obj):
        list1.append(i['name'])
    return list1
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
#top 3 characters
def convert3(obj):
    list1 = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            list1.append(i['name'])
            counter += 1
        else:
            break
    return list1
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def convertcrew(obj):
    list1 = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            list1.append(i['name'])
            break
    return list1
```

```python
def convertoverview(obj):
    l = []
    name = ''
    for i in obj:
        if i == " ":
            l.append(name)
            name = ""
        else:
            name += i
    return l;
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# remove the spces beteween the names to identify the particular person 
```
